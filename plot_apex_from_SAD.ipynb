### Plot a speficied column from APEX .SAD file for a single year, ID and vegetation type

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
# read in the file. Example uses small subset .SAD file stored in local data directory
df_sad = pd.read_csv('data/CONUNN_AGM.SAD', sep=' ', skipinitialspace=True, skiprows=8)
```

```python
# convert the date-related columns to a date string
df_sad['date_str'] =df_sad.apply(lambda x: '-'.join([str(x['Y']), str(x['M']), str(x['D'])]), axis=1)
# convert the new date string to a date column
df_sad['Date'] = pd.to_datetime(df_sad['date_str'])  # change data to yyyy-mm-dd format
```

```python
# set the desired subsets for plotting
plot_yr = 2008
plot_id = 1
plot_cpnm = 'CSPG'
plot_col = 'STL'
```

```python
# create the plot. You can change the size of the plot by changing figsize (Width, Height)
# note that it will automatically pull the year, ID, vegetation and column from the previous cell
fig, ax = plt.subplots(figsize=(12, 6))
sns.lineplot(x='Date', y=plot_col, data=df_sad[(df_sad['Y'] == plot_yr) &
                                               (df_sad['ID'] == plot_id) &
                                               (df_sad['CPNM'] == plot_cpnm)], ax=ax)
ax.set_title('ID = ' + str(plot_id) + '\n' +
             'Year = ' + str(plot_yr) + '\n' +
             'CPNM = ' + plot_cpnm)
plt.show()
```
